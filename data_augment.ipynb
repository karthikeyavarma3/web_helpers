```python
original_data = pd.read_csv(r'C:\Projects\web helpers\sentiments.csv')
```

```python
original_data.head()
```

```python
original_reviews = original_data['Text']
original_sentiments = original_data['label']
```

```python
print(original_reviews[0:5])
print(original_sentiments[0:5])
```

```python
import nlpaug.augmenter.word as naw

# Apply synonym replacement augmentation
aug = naw.SynonymAug(aug_src='wordnet')
augmented_reviews = [aug.augment(review) for review in original_reviews]
```

```python
print(augmented_reviews)
```

```python
flattened_augmented_reviews = [review for sublist in augmented_reviews for review in sublist]

print(flattened_augmented_reviews)
```

```python
list(original_reviews)
```

```python
print(len(list((original_reviews))))
print(len(list((augmented_reviews))))
```

```python
combined_reviews = list(original_reviews) + flattened_augmented_reviews
```

```python
print(len(combined_reviews))
```

```python
combined_sentiments = list(original_sentiments) + (['augmented'] * len(augmented_reviews))
```

```python
combined_data = pd.DataFrame({'text': combined_reviews, 'label': combined_sentiments} , index=range(1, len(combined_reviews) + 1))
```

```python
combined_data.shape[0]
```

```python
combined_data = combined_data.sample(frac=1).reset_index(drop=True)
```

```python
combined_data.to_csv('combined_reviews.csv', index=False)
```
